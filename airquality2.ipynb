```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print("Las 10 primeras lineas:")
print(df.head(10))
```

```python
print(f"La longitud del dataset es de : {len(df)}")
```

```python
print("Característiques de cada columna:")
print(df.info())
```

```python
max_temp = df['Temp'].max()
print(f"\nLa temperatura màxima és: {max_temp}")
```

```python
max_temp_row = df.loc[df['Temp'] == max_temp, ['Month', 'Day']]
print("\nEl mes i dia amb la temperatura màxima són:")
print(max_temp_row)
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Wind', y='Solar.R', data=df)
plt.title('Relació entre Vent i Solar.R')
plt.show()
```

```python
avg_temp_per_month = df.groupby('Month')['Temp'].mean()
avg_temp_per_month.plot(kind='bar', color='skyblue')
plt.title('Temperatura mitjana per mes')
plt.ylabel('Temperatura mitjana (°F)')
plt.show()
```
