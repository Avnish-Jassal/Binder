```python
#diezMarcas = marcasPrecio(10)
#pie (count, labels  = count, edges = 10)
# Filtrar las 10 marcas más caras
# Crear un gráfico de pastel con las 10 marcas más caras y sus precios promedio
# Agrupar por marca y obtener el precio promedio
marcasOrdenadas = df.groupby('Car Make')['Price (in USD)'].mean().sort_values(ascending=False)

# Seleccionar las 10 marcas más caras
marcasCaras = marcasOrdenadas.head(10)

# Crear un gráfico de pastel con las 10 marcas más caras y sus precios promedio
plt.figure(figsize=(8, 8))
plt.pie(marcasCaras, labels=marcasCaras.index, 
        autopct='%1.1f%%', startangle=140, wedgeprops={'edgecolor': 'black'})

print('Top Marcas de coches mas caros')
print('plt')
```
